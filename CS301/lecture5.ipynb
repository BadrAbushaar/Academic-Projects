# Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(50)
y = x**2

plt.plot(x, y)
```

```python
plt.plot(x, y, 'g--', linewidth=2, markersize=2)
```

```python
plt.plot(x, y, color='green', linewidth=1)
```

```python
fig = plt.figure(figsize=[10, 5])

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(x, y)
ax2.plot(x, y**2)
ax3.plot(x, np.sin(y))
ax4.plot(x, np.tan(y))
```

```python
fig, axes = plt.subplots(2, 3, figsize=[10, 5])
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,wspace=0.5, hspace=0.5) 
# wspace and hspace controls the percent of the figure width and figure height to use as spacing between subplots.
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].plot(x,y)
```

```python
from numpy.random import randn
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(10), 'k*', label='Data one')
ax.plot(randn(10), 'b^', label='Data two')
ax.plot(randn(10), 'r.', label='Data three')
ax.legend(loc='best')
```

```python
fig = plt.figure();
ax = fig.add_subplot( 1, 1 ,1)
ax.plot(randn(10), label= '2017')
ax.plot(randn(10), label='2018')
ax.plot(randn(10), label='2019')
ax.legend()
```
