```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset("tips")
df.columns
```

```python
df["tip"].describe()
```

```python
sns.histplot(data=df, x="tip", color="lightseagreen")
```

```python
sns.kdeplot(df.tip, shade=True)
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="kde", color="lightseagreen", fill=True) 
# kind : (scatter | kde | hist | hex | reg | resid)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex", diag_kind="hist")
```

```python
sns.boxplot(data=df.tip, orient='h')
```

```python
sns.boxplot(x="day", y="tip", orient="v", data=df, palette="rocket")
```

```python
sns.boxplot(x="tip", y="day", orient="h", data=df, palette="rocket")
```

```python
matrix_data_corr = df.corr()
matrix_data_corr
```

```python
sns.heatmap(matrix_data_corr, annot=False, cmap="Blues")
```

```python
g = sns.FacetGrid(df, col="day", row="sex")
g.map(sns.scatterplot, "total_bill", "tip", color="lightseagreen")
```
